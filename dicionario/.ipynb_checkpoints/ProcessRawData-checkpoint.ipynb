```python
import pandas as pd
```

```python
df = pd.read_csv('letras.csv')
```

```python
df = df.fillna('')
```

```python
df = df.drop_duplicates()
```

```python
df.categoria = df.categoria.apply(lambda x: x.replace('(', ''))
df.categoria = df.categoria.apply(lambda x: x.replace(')', ''))
df.categoria = df.categoria.apply(lambda x: x.strip())
df.separacao = df.separacao.apply(lambda x: x.strip())
df.silaba = df.silaba.apply(lambda x: x.strip())
df.palavra = df.palavra.apply(lambda x: x.strip())
```

```python
df.to_csv('separacao_silabica.csv')
```
